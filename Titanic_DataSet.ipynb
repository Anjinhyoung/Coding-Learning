### 라벨 인코딩(Label Encoding)
* 문자(Categorical)를 수치(Numerical)로 변환


```python
df.info()
```

```python
df.Sex.value_counts()
```

```python
def convert_sex(data):
    if data =='male':
        return 1
    elif data =='female':
        return 0
```

```python
convert_sex('male')
```

```python
df.Sex = df.Sex.apply(convert_sex)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder # 문자를 숫자로 바꿈
```

```python
le = LabelEncoder() # labelencoder를 하면 알파벳 순  C - Q - S
```

```python
df['Embarked'].value_counts()
```

```python
embarked = le.fit_transform(df['Embarked'])
embarked
```

```python
le.classes_ # 0:'C', 1:'Q' , 2:'S' , 3:'nan'
```

## 원 핫 인코딩


```python
df['Embarked_num'] = LabelEncoder().fit_transform(df['Embarked'])
```

```python
df.head()
```

```python
pd.get_dummies(df['Embarked_num'])
```

```python
df = pd.get_dummies(df,columns=['Embarked'])
```

```python
df.head()
```
