# 23.원 핫 인코딩

* 더미변수(별 의미 없는 데이터입니다)


* 원 핫 인코딩은 한 개의 요소는 True(1), 나머지 요소는 False(0)로 만들어 카테고리형  표현하는 기술



```python
blood_map={'A':0,'B':1,'AB':2,'O':3}
```

```python
df['혈액형_code'] = df['혈액형'].map(blood_map)

df.head()
```

* df['혈액형_code"]를 머신러닝 알고리즘에 넣어 데이터를 예측하려고 지시를 하면 컴퓨터는 값들간의 관계를 스스로 형성하게 됨
* 만약 B형은 1, AB형은 2라는 값을 가지고 있을 때, 컴퓨터는 'B'형 + 'AB'형 = 'O'형이라는 이상한 관계를 맺을 수 있음
* 별도의 column을 형성해주고 1개의 column에는 True, 나머지는 False라고 넣어줌으로 'A , B ,AB ,O'형의 관계는 서로 독립적이다 라는 카테고리로 표현해주는 방식

```python
df['혈액형_code'].value_counts() # 각각의 카테고리가 뭐가 있고 거기에 속하는 개수가 몇 개인지 알려줌
```

```python
pd.get_dummies(df['혈액형_code']) # 원 핫 인코딩은 문자열에서 많이 쓰임
```
