### 문제
konlpy에서 제공되는 constitution.txt를 읽어
명사를 추출하고 빈도수에 따라 워드 클라우드를 작성해보자

```python
!pip install konlpy
```

```python
from konlpy.corpus import kolaw
```

```python
text = kolaw.open('constitution.txt').read()
text
```

```python
from konlpy.tag import Okt
```

```python
okt = Okt()
noun_text = okt.nouns(text)
print(noun_text)
```

```python
noun_text.sort(key=lambda x:len(x))
print(noun_text)
```

```python
# stop_words = []
# noun_text = [each_word for each_word in noun_text if each_word not in stop_words]
# 이런 방법도 있음 단 stop_word는 불용어 등록

noun_text = [each_word for each_word in noun_text if len(each_word)>1]
print(noun_text)
```

```python
from collections import Counter
count = Counter(noun_text)

print(count)
```

```python
data = count.most_common(500)
data = dict(data)
print(data)
```
