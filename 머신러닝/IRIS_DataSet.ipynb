```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

- sepal length in cm: 꽃받침의 길이
- sepal width in cm:꽃받침의 너비
- petal length in cm:꽃잎의 길이
- petal width in cm:꽃잎의 너비

```python
data = iris.data
data
```

```python
feature_names = iris['feature_names']
feature_names
```

* ai에서 가장 많이 사용하는 자료구조는 ndarray, 데이터 프레임, 텐서

```python
import pandas as pd
```

```python
df_iris = pd.DataFrame(data, columns=feature_names) # 값,행,열 순으로 ,매개변수 행은 안 넣어도 됨 데이터 프레임에서 행은 안 넣으면 기본값으로 설정
df_iris.head()
```

```python
target = iris['target']
df_iris['target'] = target
```

```python
df_iris
```

* 데이터셋 나눌 때 나누는 세 가지 보통은 공부할 때 두 가지 train, val을 쓰지만
* 실무에서는 test까지 세 가지 씀
* train : 학습 데이터
* val : 시험 보는 데이터(모의고사)
* test : 실제로 적용(수능)

```python
from sklearn.model_selection import train_test_split
```

* 데이터로 시험 볼 때 절대로 df_iris의 프레임을 그대로 넣으면 안된다. 왜냐하면 target을 보면 0,1,2 순서대로 있는데 0으로 쏠려 있고, 1로 쏠려있고, 2로 쏠려 있어서 train데이터는 0번에서 열심히 공부하면 val에서 2번으로 시험으로 보기 때문에 랜덤으로 섞는 게 좋다.

```python
# train_test_split(독립변수, 종속변수, 테스트사이즈, 시드값...)

# train_test_split(test_size=0.2(test는 20프로,train은 80),random_state = 2023(2023으로 고정으로 섞임 2023번 섞는 게 아니라...)

X_train, X_test, y_train, y_test = train_test_split(df_iris.drop('target',axis=1), # 독립 변수 부분:train_test_split(df_iris.drop('target', 1) 데이터 프레임이 통 째로 넣어도 상관 없음 지금 열이 5개니깐 'target'만 빼면 깔끔
                 df_iris['target'], # 'target'은 종속 변수
                 test_size=0.2,
                 random_state=2023)

# X_train: df_iris의 데이터 프레임 전체에 80퍼센트
# X_test: 나머지 20프로
# y_train: X_train에 80 퍼센트에 대한 정답
# y_test: X_test의 정답
```

```python
X_train.shape,X_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
X_train
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
svc = SVC()
```

```python
svc.fit(X_train,y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
print('정답률:',accuracy_score(y_test,y_pred))
```

```python
y_pred = svc.predict([[5.9,3.0,5.1,1.8]]) # predict 안에 숫자를 조절하면 array([2])가 바뀔 수도 있고 안 바뀔 수도 있음
y_pred
```
