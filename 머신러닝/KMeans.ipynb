# 1.클러스터(Clusters)
* 유사한 특성을 가진 개체들의 집합
* 고객 분류, 유전자 분석, 이미지 분할


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=100, centers=3, random_state=2023) # 가상의 클러스터 데이터를 생성
```

```python
X = pd.DataFrame(X)
X
```

```python
y
```

```python
sns.scatterplot(x=X[0], y=X[1], hue=y)
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters = 3)
km.fit(X)
pred= km.predict(X)
```

```python
sns.scatterplot(x=X[0], y=X[1], hue=pred)
```

```python
km = KMeans(n_clusters = 5)
km.fit(X)
pred= km.predict(X)
```

```python
sns.scatterplot(x=X[0], y=X[1], hue=pred)
```

```python
# 평가값: 하나의 클러스 안에 중심점으로부터 각각의 데이터 거리를 합한 값 평균
km.inertia_
```

```python
# 엘보우 메소드
inertia_list = []

for i in range(2,11):
      km = KMeans(n_clusters= i)
      km.fit(X)
      inertia_list.append(km.inertia_)
```

```python
inertia_list
```

```python
sns.lineplot(x=range(2,11),y=inertia_list)
```

# 3.KMeans
* k개의 중심점을 찍은 후에, 이 중심점에서 각 점간의 합이 최소가 되는 중심점 k의 위치를 찾고, 이 중심점에서 가까운 점들을 중심점을 기준으로 묶는 알고리즘
* k개의 클러스터 수는 정해줘야 함


```python
inertia_list = []

for i in range(2,11):
    km = KMeans(n_clusters=i, random_state=2023)
    km.fit(ss_df)
    inertia_list.append(km.inertia_)
```

```python
inertia_list
```

```python
sns.lineplot(x=range(2,11),y=inertia_list)
```

# 4. 실루엣 스코어
* 각 군집 간의 거리가 얼마나 효율적으로 분리 돼어있는 지를 나타냄
* 실루엣 분석은 실루엣 계수를 기반으로 하는데, 실루엣 계수는 개별 데이터가 가지는 군집화 지표


```python
from sklearn.metrics import silhouette_score
```

```python
score =[]

for i in range(2,11):
    km = KMeans(n_clusters=i, random_state=2023)
    km.fit(ss_df)
    pred = km.predict(ss_df)
    score.append(silhouette_score(ss_df,pred))
```

```python
sns.lineplot(x=range(2,11),y=score)
```
