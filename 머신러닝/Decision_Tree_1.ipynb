```python
# covid
# 2020-04-01 이전:precovid
# 2020-04-01 이후 ~ 2021-04-01 이전:covid
# 2021-04-01 : postcovid

# bike_df['covid']

def covid(date):
    if str(date) < '2020-04-01': # 아스키코드로 변환해서 비교
          return 'precovid'
    elif str(date) < '2021-04-01':
          return 'covid'
    else:
          return 'postcovid'
```

```python
bike_df['date'].apply(covid)
```

```python
bike_df['covid'] = bike_df['date'].apply(lambda date:'precovid' if str(date) <'2020-04-01' else 'covid' if str(date)< '2021-04-01' else 'postcovid')
```

```python
bike_df
```

```python
# bike_df['season']
# 3월~5월: spring
# 6월~8월:summer
# 9월~11월:fall
# 12월~2월:winter

def season(month):
    if 3 <=  month  <= 5:
        return 'spring'
    elif 6 <=  month <= 8:
        return 'summer'
    elif 9 <= month <= 11:
        return 'fall'
    else:
        return 'winter'
```

```python
bike_df['month'].apply(season)
```

```python
bike_df['season'] = bike_df['month'].apply(lambda x: 'winter' if x ==12 else 'fall' if x>=9
                                            else 'summer' if x>=6 else 'spring' if x>=3 else
                                           'winter')
```

```python
bike_df[['month','season']]
```

```python
bike_df['day_night'] = bike_df['hour'].apply(lambda x: 'night' if x >= 21 else 'late evening' if x >= 19 else 'early evening' if x >= 17 else 'late afternoon' if x >= 16 else 'early afternoon' if x >= 13 else 'late morning' if x >= 11 else 'early morning' if x >= 5 else 'night')
```

```python
bike_df['day_night'] # 원 핫 인코딩 하기엔 애매해서 묶어서 함
```

```python
bike_df.drop(['datetime','month','date','hour'],axis=1,inplace=True)
```
