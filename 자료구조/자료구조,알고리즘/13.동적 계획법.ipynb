#2. 동적 계획법 알고리즘
* 피보나치 수열 n을 입력받아서 아래와 같이 계산

(https://namu.wiki/w/%ED%94%BC%EB%B3%B4%EB%82%98%EC%B9%98%20%EC%88%98%EC%97%B4)

### 2-1. 재귀호출 활용


```python
def fibonacci(num):
    if num<=1:
        return num
    return fibonacci(num-1) + fibonacci(num-2)
```

```python
fibonacci(6)
```

### 2-2 동적 계획법 활용


```python
def fibonacci(num):
    cache = [0 for index in range(num+1)]
    #print(cache)
    cache[0] = 0
    cache[1] = 1

    for index in range(2,num+1):
        cache[index] = cache[index-1] + cache[index-2]
    return cache[num]
```

```python
fibonacci(6)
```

```python
tile = [0 for index in range(1001)]
tile[1] = 1
tile[2] = 2
n = int(input())
for index in range(3,1001):
    tile[index] = tile[index-1] + tile[index-2]

print(tile[n] % 10007)
```
