# 1. 큐
* 가장 먼저 넣은 데이터를 가장 먼저 꺼낼 수 있는 구조
* 인덱스 번호가 없음
* 줄을 서는 행위와 유사
* FIFO (First-In, First-Out)

### 1-1. 큐의 사용
* 멀티테스킹을 위한 프로세스 스케줄링 방식을 구현(운영체제)
* 푸시메세지

### 1-2. 큐의 용어
* Enqueue: 큐에 데이터를 넣는 기능
* Dequeue: 큐에 데이터를 꺼내는 기능
* [VisiualGo]https://visualgo.net/en/list)

### 1-3. queue 라이브러리를 활용
* Queue(): 가장 일반적인 큐 자료구조
생성
* LifoQueue(): 나중에 입력된 데이터가 먼저 출력되는 구조 (파이썬에만 있음)
* PriorityQueue(): 데이터마다 우선순위를 넣어서 우선순위가 높은 순으로 데이터를 출력력(파이썬에만 있음)

```python
import queue

data_queue = queue.Queue()
data_queue.put("apple") # Enqueue
print(data_queue)
print(data_queue.qsize())
data_queue.put("banana") # Enqueue
print(data_queue.qsize())
print(data_queue.get()) #Dequeue
print(data_queue.qsize())
print(data_queue.get()) #Dequeue
print(data_queue.qsize())
```

```python
import queue

data_queue = queue.Queue()
data_queue.put("apple") # Enqueue
print(data_queue)
print(data_queue.qsize())
data_queue.put("banana") # Enqueue
print(data_queue.qsize())
temp=data_queue.get()
print(temp)
print(data_queue.get())
```

```python
data_queue.put('apple')
data_queue.put('김사과')
data_queue.put(160.5)
data_queue.put(True)
print(data_queue.qsize())
```

```python
for _ in range(data_queue.qsize()):
  item=data_queue.get()
  print(item)
```

```python
data_queue=queue.LifoQueue()
data_queue.put("apple")
data_queue.put("김사과")
data_queue.put(160.5)
data_queue.put(True)
print(data_queue.qsize())
print(data_queue.get())
```

```python
# 숫자가 작을수록 우선순위가 높음
data_queue=queue.PriorityQueue()
data_queue.put(10,"apple")
data_queue.put(5,"a")
data_queue.put(6,"ae")
data_queue.put(0,"e")
print(data_queue.qsize())
print(data_queue.get())
print(data_queue.get())
```

```python
### 문제
# 리스트 변수로 Queue를 enqueue,dequeue를 기능
queue_list=list()

def enqueue(data):
  queue_list.append(data)

def dequeue():
  data=queue_list[0]
  del queue_list[0]
  return data

print(len(queue_list))

enqueue(10)
enqueue(5)
enqueue(20)
enqueue(30)

queue_list  # 다 코드를 만들어서 거기다 구현하기

print(dequeue())

queue_list

print(dequeue())

queue_list
```
