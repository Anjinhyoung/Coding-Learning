### 1-1. DFS란?
* 대표적인 그래프 탐색 알고리즘
* 정점의 자식들을 먼저 탐색하는 방식
* 한 노드의 자식을 타고 끝까지 순회한 후, 다시 돌아와서 다른 형제들의 자식을 타고 내려가서 순회함

### 1-2. DFS 알고리즘 구현
* 스택과 큐 자료구조를 활용
* need_visit 스택과 visited를 활요

```python
graph = dict()

# {'A':['B', 'C']}
graph['A'] = ['B', 'C']
graph['B'] = ['A', 'D']
graph['C'] = ['A', 'G', 'H', 'I']
graph['D'] = ['B', 'E', 'F']
graph['E'] = ['D']
graph['F'] = ['D']
graph['G'] = ['C']
graph['H'] = ['C']
graph['I'] = ['C', 'J']
graph['J'] = ['I']

graph
```

```python
def dfs(graph, start_node):
    visited, need_visit = list(), list()

    need_visit.append(start_node)

    while need_visit:
        node = need_visit.pop()
        if node not in visited:
            visited.append(node)
            need_visit.extend(graph[node])
    return visited
```
