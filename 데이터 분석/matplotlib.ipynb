```python
# 여러 개의 plot을 그리는 방법
# subplot(row,column, no)

data1 = np.arange(100,200)
plt.subplot(2,1,1)  # 2행 1열 첫 번째 그래프
plt.plot(data1)

data2 = np.arange(200,300)
plt.subplot(2,1,2)  # 2행 2열 두 번째 그래프
plt.plot(data2)

plt.show()
```

```python
# 1행 3열 sublpot

data1 = np.arange(0,100)
plt.subplot(1,3,1)
plt.plot(data1)

data2 = np.arange(0,100)
plt.subplot(1,3,2)
plt.plot(data2)

data3 = np.arange(0,100)
plt.subplot(1,3,3)
plt.plot(data3)
```

```python
# 1행 3열 sublpot

data1 = np.arange(0,100)
plt.subplot(131)
plt.plot(data1)

data2 = np.arange(0,100)
plt.subplot(132)
plt.plot(data2)

data3 = np.arange(0,100)
plt.subplot(133)
plt.plot(data3)
```

# 2.스타일 옵션

```python
plt.figure(figsize=(6,8)) # inch

plt.plot([1,2,3,],[1,2,3])
plt.plot([1,2,3,],[2,4,6])

plt.title('타이틀 테스트',fontsize=30)
plt.xlabel('x축', fontsize=15)
plt.ylabel('y축', fontsize=15)

plt.show()
```

```python
plt.figure(figsize=(15,10))

plt.title('마커설정',fontsize=30)

plt.plot(np.arange(10), np.arange(10),marker='o',color='b',linestyle='')

plt.plot(np.arange(10), np.arange(10)*2,marker='v',color='r',linestyle='--')

plt.plot(np.arange(10), np.arange(10)*3,marker='*',color='y',linestyle='-.',alpha=0.4)


#범례
plt.legend(['10','10*2','10*3'],fontsize=15,loc='lower right', ncol=3)

#x축 y축 설정
plt.xlim(0,12)
plt.ylim(0,30)

plt.xticks(rotation=30)
plt.yticks(rotation=90)

# plt.grid()

plt.show()
```

```python
x = ['Math','Pro','Ds','Art','English','Korean']
y = [95,45,60,100,80,90]

plt.figure(figsize=(8,5))
plt.bar(x,y,align = 'center', alpha=0.7, color='red') #align: 글자가 앞으로 가던지 중앙으로 가던지 뒤로 가던


plt.title('subject',fontsize=25)

plt.ylabel('score of Student')

plt.show()

```

```python
x = ['Math','Pro','Ds','Art','English','Korean']
y = [95,45,60,100,80,90]

plt.figure(figsize=(8,5))
plt.barh(x,y,align = 'center', alpha=0.7, color='pink') #align: 글자가 앞으로 가던지 중앙으로 가던지 뒤로 가던


plt.title('subject',fontsize=25)

plt.xlabel('score of Student')

plt.show()

```
